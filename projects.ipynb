```python
# =====================================================================
# CODE FOR PROJECT 2: VIRAL RUMOR SIMULATOR
# =====================================================================
!pip install -q ipywidgets numpy

import time
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# 1. Setup UI Controls
gullibility_slider = widgets.IntSlider(value=40, min=10, max=90, step=5, description='Gullibility %:')
shock_value_slider = widgets.IntSlider(value=50, min=20, max=100, step=5, description='Shock Value %:')
influencer_slider = widgets.IntSlider(value=3, min=1, max=10, step=1, description='Initial Nodes:')
spread_btn = widgets.Button(description='🚨 ACTIVATE RUMOR SPREAD', layout=widgets.Layout(width='45%', height='45px'))

spread_btn.style.button_color = '#FF9500' # Operational Amber Alert
spread_btn.style.text_color = '#000000'
spread_btn.style.font_weight = 'bold'

output_area_2 = widgets.Output()

display(HTML("""
<div style='background: linear-gradient(135deg, #1C1200 0%, #000000 100%); padding: 25px; border-radius: 16px; border: 1px solid #4A3200; margin-bottom: 20px;'>
    <h2 style='color: #FFFFFF; font-family: "Segoe UI", sans-serif; font-weight: 800; margin: 0;'>🧠 SOCIAL PSYCHOLOGY RADAR CORE</h2>
    <p style='color: #FF9500; font-family: "Segoe UI", sans-serif; font-size: 13px; margin: 5px 0 0 0;'>Simulating Algorithmic Virality & Narrative Propagation Cascades Across 500 Node Interfaces</p>
</div>
"""))
display(widgets.VBox([gullibility_slider, shock_value_slider, influencer_slider, spread_btn]), output_area_2)

# 2. Network Graph Propagation Math Engine
def run_rumor_simulation(b):
    with output_area_2:
        clear_output()

        gullibility = gullibility_slider.value / 100.0
        shock = shock_value_slider.value / 100.0
        seed_count = influencer_slider.value

        # Configure Population arrays (500 nodes)
        # States: 0 = Safe/Unaware, 1 = Actively Spreading Rumor, 2 = Skeptical/Immune
        nodes_count = 500
        node_states = np.zeros(nodes_count)

        # Seed initial influencers to launch propagation wave
        initial_seeds = np.random.choice(nodes_count, size=seed_count, replace=False)
        node_states[initial_seeds] = 1

        time_steps = 30
        peak_infected_count = seed_count

        # Run generational timeline tracking loops
        for step in range(1, time_steps + 1):
            current_spreaders = np.where(node_states == 1)[0]
            if len(current_spreaders) == 0:
                break

            new_states = node_states.copy()

            # Each active carrier attempts to spread rumor to 3 random contact nodes daily
            for spreader in current_spreaders:
                targets = np.random.choice(nodes_count, size=3, replace=False)
                for t in targets:
                    if node_states[t] == 0: # If target node is currently unaware
                        # Probability threshold logic check
                        if np.random.rand() < (gullibility * shock * 1.3):
                            new_states[t] = 1 # Node catches and amplifies rumor
                        else:
                            new_states[t] = 2 # Node acts critical, builds immunity baseline

            # Spreading agents burn out or get flagged by moderation rules over time
            for spreader in current_spreaders:
                if np.random.rand() > 0.6:
                    new_states[spreader] = 2 # Move to silent/immune profile state

            node_states = new_states.copy()
            active_now = np.sum(node_states == 1)
            if active_now > peak_infected_count:
                peak_infected_count = active_now

        total_infected = np.sum(node_states == 1) + np.sum(node_states == 2) - seed_count
        viral_velocity_pct = (total_infected / nodes_count) * 100
        saturation_pct = (peak_infected_count / nodes_count) * 100

        # --- UI OUTPUT PANEL ---
        status_banner = "🔴 CRITICAL VECTOR: MASS HYSTERIA RISK DETECTED" if viral_velocity_pct > 60 else "🟢 CONTAINED NARRATIVE: LOCALIZED INCIDENT"
        theme_color = "#FF1E27" if viral_velocity_pct > 60 else "#FF9500"

        display(HTML(f"""
        <div style='font-family: "Segoe UI", sans-serif; margin-top: 15px;'>
            <div style='background-color: #030303; border: 1px solid #4A3200; border-radius: 16px; padding: 25px; max-width: 650px; color: #FFFFFF;'>
                <h3 style='color: {theme_color}; margin: 0 0 15px 0; font-size: 15px; font-weight: 900;'>{status_banner}</h3>

                <div style='margin-bottom: 15px;'>
                    <div style='display: flex; justify-content: space-between; font-size: 13px; margin-bottom: 4px;'>
                        <span><b>Total Network Penetration Rate:</b></span>
                        <span style='color: #FF9500; font-weight: bold;'>{int(total_infected)} Users Infected ({viral_velocity_pct:.1f}%)</span>
                    </div>
                    <div style='background-color: #1A1200; width: 100%; height: 8px; border-radius: 4px; overflow: hidden;'>
                        <div style='background: #FF9500; width: {viral_velocity_pct}%; height: 8px;'></div>
                    </div>
                </div>

                <div style='margin-bottom: 20px;'>
                    <div style='display: flex; justify-content: space-between; font-size: 13px; margin-bottom: 4px;'>
                        <span><b>Peak Simultaneous Content Spreading Wave:</b></span>
                        <span style='color: #FFFFFF; font-weight: bold;'>{int(peak_infected_count)} Concurrent Accounts ({saturation_pct:.1f}%)</span>
                    </div>
                    <div style='background-color: #222; width: 100%; height: 8px; border-radius: 4px; overflow: hidden;'>
                        <div style='background: #FFFFFF; width: {saturation_pct}%; height: 8px;'></div>
                    </div>
                </div>

                <div style='background-color: #0A0600; border: 1px solid #4A3200; padding: 15px; border-radius: 8px; font-size: 13px; line-height: 1.6; color: #CCCCCC;'>
                    📋 <b>Network Behavior Log Stream:</b><br>
                    • Outbreak metrics run across <b>30 dynamic contact clock intervals</b>.<br>
                    • Network Density State: High clustering amplification observed.<br>
                    • Saturation Profile: {f"Rumor successfully breached community tracking filters, entering viral trend metrics loops." if viral_velocity_pct > 50 else "Skepticism thresholds checked the amplification loop safely, causing early information decay."}
                </div>
            </div>
        </div>
        """))

spread_btn.on_click(run_rumor_simulation)
```

```python
# =====================================================================
# CODE FOR PROJECT 3: GENERATIVE GLITCH ART CANVAS
# =====================================================================
!pip install -q ipywidgets numpy PIL

import io
import base64
import numpy as np
import ipywidgets as widgets
from PIL import Image
from IPython.display import display, HTML, clear_output

# 1. Setup Control UI Elements
symmetry_axis = widgets.SelectionSlider(options=['Horizontal Grid', 'Vertical Grid', 'Quad Matrix Axis'], value='Quad Matrix Axis', description='Symmetry Setup:')
glitch_factor = widgets.IntSlider(value=30, min=0, max=100, step=5, description='Glitch Noise %:')
color_theme = widgets.Dropdown(options=['Cyberpunk Neon', 'Muted Vaporwave', 'Monochrome Terminal'], value='Cyberpunk Neon', description='Palette Token:')
render_art_btn = widgets.Button(description='🔮 RE-GENERATE CANVAS MATRICES', layout=widgets.Layout(width='45%', height='45px'))

render_art_btn.style.button_color = '#A333FF' # Neon Creative Purple
render_art_btn.style.text_color = '#FFFFFF'
render_art_btn.style.font_weight = 'bold'

output_area_3 = widgets.Output()

display(HTML("""
<div style='background: linear-gradient(135deg, #12001A 0%, #000000 100%); padding: 25px; border-radius: 16px; border: 1px solid #4A0080; margin-bottom: 20px;'>
    <h2 style='color: #FFFFFF; font-family: "Segoe UI", sans-serif; font-weight: 800; margin: 0;'>🎨 NEURAL COMPOSER CONSOLE</h2>
    <p style='color: #B07FFF; font-family: "Segoe UI", sans-serif; font-size: 13px; margin: 5px 0 0 0;'>Algorithmic Generative Art Studio Mapping Parametric Waves & Pixels Distortions Matrices</p>
</div>
"""))
display(widgets.VBox([symmetry_axis, glitch_factor, color_theme, render_art_btn]), output_area_3)

# 2. Computational Canvas Matrix Rendering Engine
def run_art_generation(b):
    with output_area_3:
        clear_output()

        sym_mode = symmetry_axis.value
        noise_level = glitch_factor.value / 100.0
        theme = color_theme.value

        # Build raw pixel workspace (Width 300px x Height 150px)
        w, h = 300, 150
        pixel_array = np.zeros((h, w, 3), dtype=np.uint8)

        # Math iteration loops painting algorithmic gradients via sine/cosine waves
        for y in range(h):
            for x in range(w):
                # Calculate coordinates centered around axis points
                cx = x - w/2
                cy = y - h/2

                # Apply structural symmetry formulas
                if sym_mode == 'Horizontal Grid':
                    cy = abs(cy)
                elif sym_mode == 'Vertical Grid':
                    cx = abs(cx)
                elif sym_mode == 'Quad Matrix Axis':
                    cx, cy = abs(cx), abs(cy)

                # Equation metrics computing mathematical wave patterns
                wave_val = np.sin(cx / 12.0) * np.cos(cy / 12.0)
                int_val = int((wave_val + 1.0) * 127.5)

                # Assign distinct color channels according to palette selections
                if theme == 'Cyberpunk Neon':
                    pixel_array[y, x] = [int_val, 0, 255 - int_val] # Intense Magenta / Cyan profile
                elif theme == 'Muted Vaporwave':
                    pixel_array[y, x] = [255 - int_val//2, 180, int_val] # Soft Pink / Purple hues
                else:
                    pixel_array[y, x] = [0, int_val, 0] # Matrix Terminal Amber-Green glow

        # Apply Glitch-Noise Bitwise Corruption Layer if requested
        if noise_level > 0.0:
            noise_mask = np.random.rand(h, w) < noise_level
            # Inject randomized color fragmentation blocks across mask index arrays
            pixel_array[noise_mask] = np.random.randint(0, 255, size=(np.sum(noise_mask), 3))

        # Convert raw NumPy data array safely into image byte assets via PIL
        img = Image.fromarray(pixel_array)
        img_buffer = io.BytesIO()
        img.save(img_buffer, format="JPEG")
        encoded_canvas = base64.b64encode(img_buffer.getvalue()).decode('utf-8')

        # --- UI OUTPUT PANEL ---
        display(HTML(f"""
        <div style='font-family: "Segoe UI", sans-serif; margin-top: 15px;'>
            <div style='background-color: #020205; border: 1px solid #4A0080; border-radius: 16px; padding: 25px; max-width: 650px; color: #FFFFFF;'>
                <div style='text-align: center; margin-bottom: 20px;'>
                    <img src="data:image/jpeg;base64,{encoded_canvas}" style='width: 100%; max-width: 500px; border-radius: 8px; border: 2px solid #4A0080; box-shadow: 0 0 20px #A333FF33;' />
                </div>
                <hr style='border: 0; border-top: 1px solid #4A0080; margin: 15px 0;'>
                <div style='display: grid; grid-template-columns: 1fr 1fr; gap: 15px; font-size: 13px;'>
                    <div>🎨 <b>Canvas Dimensions:</b> {w} x {h} Pixels</div>
                    <div>🧬 <b>Symmetry Mode Deployed:</b> {sym_mode}</div>
                    <div>🌌 <b>Calculated Glitch Density index:</b> {noise_level*100:.0f}% Allocation</div>
                    <div>🔮 <b>Studio Pipeline Render Status:</b> Success Complete</div>
                </div>
            </div>
        </div>
        """))

render_art_btn.on_click(run_art_generation)
```
